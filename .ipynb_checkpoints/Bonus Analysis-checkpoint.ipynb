```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import matplotlib.pyplot as plt
from config import username, password
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/sql-challenge')
connection = engine.connect()
```

```python
query1 = pd.read_sql('select * from salaries', connection)
query2 = pd.read_sql('select e.emp_no, s.salary, t.title from employees as e join salaries as s on s.emp_no = e.emp_no join Titles as t on t.title_id = e.emp_title_id',connection)
```

```python
df1 = query1
df1['salary'].fillna(0)
```

```python

bins = np.arange(0,125000,5000)
df1['range'] = pd.cut(df1['salary'], bins=bins)
df1.groupby('range').emp_no.count().hist()
plt.title('Histogram of Salary ranges of Employees')
plt.xlabel('Salary range')
plt.ylabel('Number of Employees')
plt.show()
# Hence maximum salary is in the range 0-10000 (approx 15 employees)
```

```python
df2 = query2.groupby('title').mean()
df2.plot(kind = 'bar', rot = 45, title = 'Average Salary by Title', xlabel = 'Employee Titles', ylabel = 'Average Employee Salaries ($)')
plt.show()
```
